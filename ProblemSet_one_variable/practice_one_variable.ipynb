```r
options(repr.plot.width=20, repr.plot.height=3)
ggplot(data=diamonds, aes(diamonds$price)) + 
  geom_histogram(binwidth=1, col="black",fill="grey") + 
  labs(title="Histogram for diamond prices") +
  labs(x="Price", y="Count")+
scale_x_continuous(limits = c(0, 2000), breaks = seq(0, 2000, 50))+
coord_cartesian(xlim = c(1400,1600))
ggsave("text.png",width=25, height=10)
```

# Frequency polygon 

```r
ggplot(aes(x = carat), data = diamonds) + 
  geom_freqpoly(aes(color = color), binwidth=.1) + 
  scale_x_continuous(limits = c(0, 6), breaks = seq(0, 6, 1)) + 
  xlab('Carat') + 
  ylab('Count')
```
