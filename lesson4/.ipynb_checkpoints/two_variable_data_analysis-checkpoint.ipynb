# Analyisis of two continoius variables

```r
pf<-read.csv("C:/Users/sleeba/OneDrive/Code Base - R/Introduction to R -- Udacity/lesson3/pseudo_facebook.tsv",
            sep="\t",header=T)
library(ggplot2)
```

We'll play around with **`scatter plots`** 

```r
#qplot(x=age,y=friend_count, data = pf)
```

```r
summary(pf$age)
```

```r
ggplot(aes(x=age,y=friend_count),data=pf)+
geom_point()+
xlim(13,90)
```

This is overplotted in lower range of the plot. So we bring in **`alpha`** parameter in and set the transperancy to see the data_points better.

```r
ggplot(aes(x=age,y=friend_count),data=pf)+
geom_point(alpha=1/20)+
xlim(13,90)
# alpha is 1/20 means one dot is equivalent to 20 data points
```

Using **`geom_jitter`** we can add some noise to edges and get a better picture

```r
ggplot(aes(x=age,y=friend_count),data=pf)+
geom_jitter(alpha=1/20)+
xlim(13,90)
```
